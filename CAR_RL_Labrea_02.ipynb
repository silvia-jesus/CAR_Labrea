Não há elemento de identificação entre Reserva Legal e Área do Imóvel, apenas a informação espacial (sobreposição das camadas). Sendo assim, será gerado o centroide dos polígonos de RL e então a identificação com os polígonos dos imóveis.

O processamento identificado acima não funcionou visto que há sobreposição de imóveis.

```python
start = time.time()

points_RL = rl_geo.copy()
points_RL['geometry'] = points_RL['geometry'].representative_point() #gera o centroide no interior do poligono
points_RL.head()

end = time.time()
duration = end-start

print(f'O processamento demorou {duration} segundos.')
print(points_RL.shape)
```

```python
start = time.time()

pts_RL_id = gpd.overlay(points_RL, im_geo, how='identity')


end = time.time()
duration = end-start
print(f'O processamento demorou {duration} segundos.')
print(pts_RL_id.shape)

pts_RL_id.head()
```

```python
condicao = [(im_geo_diss.NUM_MODULO	>4),(im_geo_diss.NUM_MODULO	<=4) & (im_geo_diss.NUM_MODULO	>2),(im_geo_diss.NUM_MODULO	<=2) & (im_geo_diss.NUM_MODULO	>1),(im_geo_diss.NUM_MODULO	<=1)]
classes = ['IV','III','II','I']

im_geo_diss['Classe_MF'] = np.select(condicao, classes)

im_geo_diss.head()

```

```python
start = time.time()
n=im_geo_diss.shape[0] #save each polygon in a different file in its respective category directory
for i in range(n):
    selection = im_geo_diss.iloc[[i]]
    selection.reset_index(drop=False, inplace=True)
    municipio=selection.iloc[0]['NOM_MUNICI']
    name=selection.iloc[0]['COD_IMOVEL']
    mf=selection.iloc[0]['Classe_MF']
    if not os.path.exists(f'/content/drive/MyDrive/CAR_RL_Labrea/OUTPUT/IMOVEL_SPLIT/{mf}'): #separa em pastas diferentes para não carregar a memória do Google Drive
      os.makedirs(f'/content/drive/MyDrive/CAR_RL_Labrea/OUTPUT/IMOVEL_SPLIT/{mf}')
    Path('./').mkdir(parents=True, exist_ok=True)

    selection.to_file(f'/content/drive/MyDrive/CAR_RL_Labrea/OUTPUT/IMOVEL_SPLIT/{mf}/IM_{municipio}_{name}.shp')


end = time.time()
duration = end-start
print(f'O processamento demorou {duration} segundos.')
```
